##CLEAN UP THE DATA

drop naN

```python
nan_df = all_months_data[all_months_data.isna().any(axis=1)]
nan_df.head()
 
all_months_data = all_months_data.dropna(how='all')
all_months_data
```

remove 'Or' 

we have to add the month column here first then remove the data // converted the data to int after removing 'Or'

```python
all_months_data.loc[:,'Month'] = all_months_data['Order Date'].str[0:2]

all_months_data = all_months_data[all_months_data['Month'].str[0:2] != 'Or']
all_months_data['Month'] = pd.to_numeric(all_months_data['Month'])
all_months_data.head()
```

MAKE ALL THE STRING(ORDER , PRICE) ITEMS TO INTEGER 

```python
all_months_data['Price Each'] = pd.to_numeric(all_months_data['Price Each'])
all_months_data['Quantity Ordered'] = pd.to_numeric(all_months_data['Quantity Ordered'])
all_months_data.head(50)
```

AUGMENT DATA WITH ADDITIONAL COLUMN

ADD MONTH COLUMN

Question 1:  WHAT IS THE BEST MONTH FOR SALE ? HOW MUCH WAS EARNED DURING THAT MONTH?

steps -> Add  sales column

```python
all_months_data['Sales'] = all_months_data['Price Each'] * all_months_data['Quantity Ordered']
all_months_data
```

```python
results = all_months_data.groupby('Month').sum()
results.head(2)
```

```python
months = ['JAN','FEB','MAR','APR','MAY','JUN','JUL','AUG','SEP','OCT','NOV','DEC']
plt.bar(months,results['Sales'])
plt.xticks(months)
plt.ylabel('Sales in USD($)')
plt.xlabel('Months')
plt.show()
```

QUESTION -> WHAT CITY HAD HIGHEST NUMBER OF SALES?

```python
## we will get the city first

def get_city(address):
    return address.split(',')[1]
def get_state(address):
    return address.split(',')[2].split(' ')[1]
all_months_data['City'] = all_months_data['Purchase Address'].apply(lambda x:f"{get_city(x)} ({get_state(x)})")
all_months_data
```

```python
city = all_months_data.groupby('City').sum()
city.head()
```

```python
cities = [city for city, df in all_months_data.groupby('City')]
plt.bar(cities,city['Sales'])
plt.xticks(cities, rotation = 'vertical',size =8 )
plt.ylabel('Sales in USD($)')
plt.xlabel('City ')
plt.show()
```

QUESTION : WHAT TIME SHOULD WE DISPLAY ADVERTISEMENTS TO MAXIMIZE LIKELIHOOD OF CUSTOMER'S BUYING PRODUCT

```python
all_months_data['Order Date'] = pd.to_datetime(all_months_data['Order Date'])

all_months_data['Hour'] = all_months_data['Order Date'].dt.hour
all_months_data['Minutes'] = all_months_data['Order Date'].dt.minute

all_months_data
```

```python
hours =  [hour for hour , df in all_months_data.groupby('Hour')]

plt.plot(hours,all_months_data.groupby(['Hour']).count())
plt.xticks(hours)
plt.ylabel('Number of Orders')
plt.xlabel('Hours')
plt.grid()
plt.show()
```

QUESTION : WHAT PRODUCTS ARE SOLD TOGETHER ?

```python
df= all_months_data[all_months_data['Order ID'].duplicated(keep=False)].copy()
df.loc[:,'Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x : ','.join(x))
df.head(4)
```

now we will reomve these duplicated items


```python
df = df[['Order ID', 'Grouped']].drop_duplicates()
df
```

```python
from itertools import combinations 
from collections import Counter

count = Counter()

for row in  df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list,2)))

print(count.most_common())
```
